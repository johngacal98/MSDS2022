# Introduction

## Overview

Cigarette smoking has contributed to the deterioration of health outcomes of Filipinos. Globally, the World Health Organization (WHO) has posited that 80% of worldwide smokers come from developing countries such as the Philippines. The health governing body has also reported that around 87,600 Filipinos die annually due to tobacco-related diseases ([WHO, 2020](#who1)). Not only does smoking cause harm to the user, it also generates negative externalities towards other people. 14% of smoking-related deaths worldwide are caused by exposure to secondhand smoke from cigarettes ([WHO, 2020](#who1)). 
Moreover, increased amounts of smoking activity has reduced an individual’s consumption set for all other necessities. A report by the American Cancer Society argues that Filipinos would have to spend 6.45% of their income yearly so that they can satiate their need to smoke daily ([Drope, et al., 2018](#drope)). This makes it difficult for poorer households, who have members with a higher likelihood of smoking, to rise out of poverty. 

Likewise, the Philippines’ Department of Health (DOH) suggests that the Philippines spends Php 188 billion yearly on hospitalization expenditures, medicine, and all other forms of treatment that are associated with only four out of 40 known smoking-related diseases - namely lung cancer, chronic obstructive pulmonary disease, heart disease, and stroke ([DOH, n.d.](#doh)). 
Smoking also has generated adverse environmental effects in the Philippines. Proper cigarette butt disposal is not evident in most public areas in the country. There are around 37,000 tons of cigarette butts and packaging that end up in different dump sites around the Philippines each year ([DOH, n.d.](#doh)). When cigarette butts are not disposed of properly, it contributes to land, air, and water pollution which stems from toxic chemicals emitted by such waste. With all the social, economic, and environmental repercussions that the practice of cigarette smoking has generated in the Philippines, it is imperative that the government must put into priority major initiatives and programs that would curb down the propensity of Filipinos to consume cigarette products. 

There have been laws and public policies enacted by the Philippine government in order to disincentivize smoking behavior. Former president Gloria Macapagal-Arroyo has passed the Republic Act 9211, which regulates the packaging, use, sale, and the distribution of tobacco products. Major guidelines enacted in the law include the prohibition of smoking in public places such as universities, recreational facilities, hospitals, and public transportation ([Tobacco Regulation Act of 2003](#9211)). Sale of tobacco products to minors aged eighteen years of age and below was also prohibited, and advertisements of tobacco products must always indicate a disclaimer that smoking is dangerous to one’s health. Cigarette packages were required to have printed text warnings that enumerate the risks of smoking. Although the law took into account the welfare of Filipinos, it also took into account the tradeoffs of the policy to different stakeholders in the tobacco industry. While smoking incidence throughout the Arroyo administration showed signs of decline, it remained to be persistently high. 30.6% and 27.6% of Filipinos aged 15 years and above were regular smokers in 2005 and 2010, respectively ([The World Bank, n.d.](#wb)). 
    
The 15th Congress of the Philippines, coinciding with Benigno Aquino’s presidency, passed RA 10351, or An Act Restructuring the Excise Tax on Alcohol and Tobacco Products - a policy that restructures the excise tax on alcohol and tobacco products ([Republic Act 10351, 2012](#excise)). The policy led to the imposition of an ad-valorem tax equal to 20 percent of the net retail price on all cigarette products, excluding the previously imposed value-added tax to consumer goods. It also added a specific tax, initially set at five pesos per cigarette stick, which increases every year after 2013 by 4%. 

Concurrently, the same administration passed the Republic Act 10643, popularly known as the Graphic Health Warnings Law, to fully inform every Filipino on the health repercussions of smoking, and to protect Filipinos from the adverse interests of the stakeholders in the tobacco industry ([Republic Act 10643, 2012](#graphic)). Among its implementing guidelines include the printing of a photographic warning of the possible effects of smoking to an individual’s health, which is accompanied by an appropriate textual warning that explains the visual representation. This graphic warning must cover 50% of each side of the cigarette box. Likewise, cigarette boxes are now forbidden to put descriptors such as ‘low tar’, ‘mild’, ‘light’, or ‘ultra light’ which may lead to erroneous public perception that some cigarettes are less harmful compared to other brands or types. Lawmakers recognized that text warnings have been proven to be empirically insufficient in cascading to individuals the adverse effects of smoking in a person’s health. An empirical study made by the WHO, which was cited by the law, argues that pictorial and graphic warnings tend to be more effective in enlightening individuals on the dangers of habitual cigarette smoking. During the Aquino administration, smoking incidence among Filipinos aged fifteen years and above have decreased by an average of 0.60% yearly ([An Act Restructuring the Excise Tax on Alcohol and Tobacco Products, 2012](#excise)). 

Recently in 2019, during the Duterte Administration, the Universal Health Care Act ([Republic Act No. 11223](#uhc)) was enacted to improve health outcomes of Filipinos and prevent non-communicable diseases - which includes tobacco-related diseases. With the persistent incidence of tobacco-related diseases as mentioned above, this study seeks to help policymakers prioritize programs that will minimize tobacco consumption among Filipinos.

```python
# load the GATS 2009 Survey Data 
gats_2009 = pd.read_csv('GATS 2009 Survey Data.csv', index_col=0,
                        low_memory=False)
gats_2009['2015'] = gats_2009.apply(lambda row: 0, axis=1)
gats_2009['b06a'] = pd.to_numeric(gats_2009['b06a'], errors='coerce')
gats_2009 = gats_2009[gats_2009['f01d'] != 'refused']

# count number of smokers in the entire household, including respondent
data = gats_2009[['hh4e001', 'hh4e002', 'hh4e003', 'hh4e004', 'hh4e005',
                  'hh4e006', 'hh4e007', 'hh4e008', 'hh4e009', 'hh4e010']]
data = data.replace({2: 0, 7: 0})
data['smoker_count_in_hh'] = data.sum(axis=1)
gats_2009 = pd.merge(gats_2009, data['smoker_count_in_hh'],
                     left_index=True, right_index=True)

# for respondents that answered the number of cigarettes they smoked weekly
# instead of daily we divided their response by seven
def week_smokers(col1, col2):
    """Return daily intake for weekly frequencies of cigarette consumption."""
    if col1 == 'r smokes product but not every day' or col1 == 888:
        return col2 / 7
    else:
        return col1

gats_2009['daily_cigarette_intake'] = gats_2009.apply(lambda x:
                                                      week_smokers(x['b06a'], 
                                                                   x['b06a1']),
                                                      axis=1)

# gats 2009 cigarette price per stick
cols = ['f01b', 'f01c', 'f01d']
gats_2009[cols] = gats_2009[cols].fillna(1)
gats_2009['f01d'] = pd.to_numeric(gats_2009['f01d'], errors='coerce')
gats_2009['f02'] = pd.to_numeric(gats_2009['f02'], errors='coerce')
gats_2009['cigarette_price'] = gats_2009['f02'] / (gats_2009['f01b'] * 
                                                   gats_2009['f01c'] * 
                                                   gats_2009['f01d'])
```

## Applying statistical methods 

### Two Proportion Z-Test

This statistical test was employed to make an assertion about whether the proportions of one population are the same as another. The assumptions needed to be able to use this test include independence, randomly sampled from the population, and the sample must reflect the underlying distribution ([Stasiak, K. (2020)](#ztest). The hypothesis test used are as follows:

- Ho:  proportions of smokers in 2009 and 2015 are the same.
- Ha:  proportions of smokers in 2009 and 2015 are different.

In this study, we used this test to see if the difference between the proportion of smokers in 2009 and 2015 is statistically significant. 


### Mann-Whitney U test

Mann-Whitney U test (also called Wilcoxon rank-sum test) is a nonparametric test that is used to test whether two samples are likely to derive from the same population (i.e., that the two populations have the same distribution). It compares whether there is a difference in the target variable for the two independent groups. Thus, it is applied in the same data situation as t-test or an ANOVA for independent samples. However, the Mann-Whitney U test is used when the data are either non-normally distributed or the measurement scale of the dependent variable is not ordinal.

We applied this test to see if there is a significant difference between the median number of cigarettes being smoked by active smokers between 2009 and 2015.

The hypothesis test, as applied to our study, is stated as follows:

- Ho: The two samples that represent the Philippine population of cigarette smokers in 2009 and in 2015, have the same distribution of daily cigarette consumption. This implies that the median cigarette consumption of Filipino smokers between 2009 and 2015 are the same.
- Ha: The two samples that represent the Philippine population of cigarette smokers in 2009 and in 2015, do not have the same distribution of daily cigarette consumption. To reject the null hypothesis is to say that the population distribution of the cigarette consumption of Filipino smokers between 2009 and 2015 are different. 


### Kruskal-Wallis H-test

The Kruskal-Wallis test is one of the non-parametric tests that is used as a generalized form of the Mann Whitney U test. The hypothesis for this test is stated as follows:

- Ho: the median or mean ranks of the groups being compared are the same.
- Ha:  the median or the mean ranks of the groups being compared are not the same.

The null hypothesis tests whether ‘k’ number of samples has been drawn from the same/identientical population with the same/identical median. This test assesses the difference against average ranks in order to determine whether or not they are likely to come from samples drawn from the same population. It may be noted that there are certain assumptions in the Kruskal-Wallis test such as independence of observations, observations must be drawn from the population by random sampling, and the distribution of the data across all groups would follow the same shape and dispersion.  The last assumption of the Kruskal-Wallis test would be assessed using Levene’s test for the equality of variances of the data across different groups. Noting that the GATS data was generated through multi-cluster stage sampling, the second assumption is fulfilled. For this study, we assumed the independence of observations that are recorded in the survey. We will utilize this non-parametric version of One-Way ANOVA to determine if there are significant differences in the median amount of cigarette consumption of individuals across different income classifications and educational attainment.

### Conover-Ivan Post-Hoc Test

A rejection of the null hypothesis of the Kruskal-Wallis test only allows us to conclude that there are at least two groups that have different medians or mean ranks. The Kruskal-Wallis test does not tell us which among the groups have significantly different medians. In order to identify which groups differ, post-hoc hypothesis testing methods were used. The p-values generated from these post-hoc hypothesis testing methods are controlled using Holm’s Step Down Procedure, in order to ensure that we have not falsely rejected too many null hypotheses (i.e. reducing Type I error). The Conover-Ivan Post-Hoc test is available through the `scikit_posthocs` library.

### Logistic Regression

Logistic regression is a Machine Learning classification algorithm that is used to predict the probability of a categorical dependent variable. The independent variables are linearly related to log odds. In this study, we will use Logistic regression to estimate the odds ratio for the event that a smoker considers quitting given their exposure to health warnings across different media. By analyzing the odds ratios,  we can identify what are the features that could increase or decrease the odds of a smoker trying to quit smoking. 

For inference purposes, we checked the assumptions of the logistic regression model before fitting our data. Since we have 9,582 observations from the model and the number of smokers trying to quit and not quit are generally balanced, we can conclude that the large sample size requirement is met. Another requirement that we need to check before implementing logistic regression is the presence of multicollinearity within the explanatory features of the model. To do this, we calculated the Variance Inflation Factor (VIF) of each variable and checked whether any variable causes multicollinearity in the model

**Z-test for proportions**

```python
from statsmodels.stats.proportion import proportions_ztest

# Ho : 2009 = 2015
# Ha: 2009 != 2015

significance = 0.025

smoker_2009, sample_size_2009 = (2751, 9695)
smoker_2015, sample_size_2015 = (2799, 11644)

successes = np.array([smoker_2009, smoker_2015])
samples = np.array([sample_size_2009, sample_size_2015])

# compute the stat and pvalue using 2sided test
stat, p_value = proportions_ztest(count=successes, nobs=samples,
                                  alternative='two-sided')

# print the results
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))
if p_value > significance:
   print ("Fail to reject the null hypothesis")
else:
   print ("Reject the null hypothesis :"
          " Proportion of smokers in 2009 != 2015")
```

**Kruskal-Wallis H-Test** <br>

To determine whether the median across the income brackets are statistically different from each other, we used the Kruskal-Wallis test. This is the non-parametric equivalent of a One-Way ANOVA test. 

One of the assumptions for this test is that the distribution of the data must be the same across different groups. To evaluate the compliance of the data to the following assumption, we have used Levene's test to test whether the variances across groups are similar. 

For the Levene's test, the null hypothesis posits that the variances across all groups are equal. On the other hand, the alternative hypothesis argues that the variances across all groups are not equal.

```python
# get relevant data
data = df[(df['smoker'] == 1) & (df['daily_cigarette_intake'] <= 31.5)]
data = data.drop(data.columns[0], axis=1)
data = data[['2015', 'income_bracket', 'highest_educ_attainment',
             'daily_cigarette_intake']]
income_rep = {'income_bracket':
             {'no income': 'no income or undisclosed',
              'refused': 'no income or undisclosed',
              "don't know": 'no income or undisclosed'}}
data = data.replace(income_rep)

# get cigarette intake per income bracket
first_bracket = data[data['income_bracket'] == 'no income or undisclosed']
second_bracket = data[data['income_bracket'] == '1 to 3,499']
third_bracket = data[data['income_bracket'] == '3,500 to 4,999']
fourth_bracket = data[data['income_bracket'] == '5,000 to 8,499']
fifth_bracket = data[data['income_bracket'] == '8,500 to 20,999']
sixth_bracket = data[data['income_bracket'] == '21,000 or higher']

first_bracket_smoke = first_bracket['daily_cigarette_intake'].to_numpy()
second_bracket_smoke = second_bracket['daily_cigarette_intake'].to_numpy()
third_bracket_smoke = third_bracket['daily_cigarette_intake'].to_numpy()
fourth_bracket_smoke = fourth_bracket['daily_cigarette_intake'].to_numpy()
fifth_bracket_smoke = fifth_bracket['daily_cigarette_intake'].to_numpy()
sixth_bracket_smoke = sixth_bracket['daily_cigarette_intake'].to_numpy()

# perform Levene's test to check whether the variances are equal or not
test_stat, p_value = levene(first_bracket_smoke, second_bracket_smoke, 
                            third_bracket_smoke, fourth_bracket_smoke,
                            fifth_bracket_smoke, sixth_bracket_smoke)
print('Levene Test Statistic: ', test_stat)
print('p-value: ', p_value)
```

From the result above, we do not have enough evidence at the 5% level of significance to reject the null hypothesis. Hence, the variances across all groups are equal and we proceed with the implementation of the Kruskal-Wallis test.

```python
# implement the Kruskal-Wallis H-test

H_stat, p_value = kruskal(first_bracket_smoke, second_bracket_smoke, 
                          third_bracket_smoke, fourth_bracket_smoke,
                          fifth_bracket_smoke, sixth_bracket_smoke)
print('Kruskal-Wallis H-statistic: ', H_stat)
print('p-value: ', p_value)
```

Based on the Kruskal-Wallis test, we have enough evidence to reject the null hypothesis that the median consumption of cigarettes are similar across different income brackets. This confirms that the difference between some of the medians are statistically significant. 

To identify though which among the income bracket groups have different median daily cigarette consumption, we need to perform a post-hoc test after rejecting the null hypothesis of a Kruskal-Wallis test.

We used the Conover-Ivan hypothesis test, correcting the p-values using Holm's method that is built-into the `posthoc_conover` function of the `scikit-posthocs` library.

```python
# Conover-Ivan post-hoc test

post_hoc_inc = sp.posthoc_conover(data, val_col='daily_cigarette_intake', 
                                   group_col='income_bracket', 
                                   p_adjust='holm')
sns.heatmap(post_hoc_inc, annot=True, fmt='.2f', cmap='OrRd_r')
plt.title('Figure 9: Heatmap of the p-values for the Pairwise Comparisons '
          'Across Income Brackets\n\n', 
          fontdict= {'fontsize': 12, 'fontweight':'bold'});
```

Based on the plot above, we see that the median consumption of the smokers who earn the least - those with income brackets at `no income or undisclosed` and `1 to 3,499` - are significantly different from the higher-income brackets. There appears no significant difference in the median consumption among and between the higher income brackets, from `3,500 to 4,999` and above.

This difference in median consumption may be attributed to the increase in tax prices on cigarettes. This may shed light on the price-sensitivity of lower income smokers, as they respond more strongly to price increases given reduced resources for other necessities [(Choi, 2016)](#choi).

Hence, we used Kruskal-Wallis to compare the medians of smoking consumption between groups of educational attainment.

Before proceeding with the Kruskal-Wallis test, we first checked whether the equality of variances assumption is fulfilled using Levene's test.

```python
# filter relevant data
educ_rep = {'highest_educ_attainment':
           {'no formal schooling': 'no formal schooling or undisclosed',
            'no grade completed': 'no formal schooling or undisclosed',
            'refused': 'no formal schooling or undisclosed',
            'elementary undergraduate': 'elementary undergrad or grad',
            'elementary graduate': 'elementary undergrad or grad',
            'high school graduate': 'HS undergrad or grad',
            'high school undergraduate': 'HS undergrad or grad',
            'post secondary': 'post secondary undergrad or grad',
            'post secondary, includes in yrs 1, 2, or 3': \
            'post secondary undergrad or grad',
            'college undergraduate': 'college undergrad or grad',
            'college graduate': 'college undergrad or grad',
            'post graduate degree completed': \
            'post secondary undergrad or grad'}
           }
data = data.replace(educ_rep)

# get cigarette intake per educational attainment level
elem = data[data['highest_educ_attainment'] == 'elementary undergrad or grad']
hs = data[data['highest_educ_attainment'] == 'HS undergrad or grad']
college = data[data['highest_educ_attainment'] == 'college undergrad or grad']
post_secondary = data[data['highest_educ_attainment'] == \
                      'post secondary undergrad or grad']
no_schooling = data[data['highest_educ_attainment'] == \
                    'no formal schooling or undisclosed']
elem_smoke = elem['daily_cigarette_intake'].to_numpy()
hs_smoke = hs['daily_cigarette_intake'].to_numpy()
college_smoke = college['daily_cigarette_intake'].to_numpy()
post_secondary_smoke = post_secondary['daily_cigarette_intake'].to_numpy()
no_schooling = no_schooling['daily_cigarette_intake'].to_numpy()

# perform Levene's test
test_stat, p_value = levene(elem_smoke, hs_smoke, college_smoke, 
                            post_secondary_smoke, no_schooling)
print('Levene Test Statistic: ', test_stat)
print('p-value: ', p_value)
```

From the p-value calculated above, we reject the null hypothesis that the variances of daily cigarette consumption across all the highest educational attainment levels are equal. Hence, the equal variances assumption of the Kruskal-Wallis test is violated and therefore the resulting H-statistic and the p-value of the subsequent hypothesis test performed must be taken with some grain of salt.

Nevertheless, we proceed with the implementation of the Kruskal-Wallis H-statistic to validate the insight that we have visualized after doing EDA.

```python
H_stat, p_value = kruskal(elem_smoke, hs_smoke, college_smoke, 
                          post_secondary_smoke, no_schooling)
print('Kruskal-Wallis H-statistic: ', H_stat)
print('p-value: ', p_value)
```

Based on the Kruskal-Wallis test, we have enough evidence to reject the null hypothesis (at the significance level of 0.05) that the median consumption of cigarettes are similar across different income brackets. This confirms that the difference between some of the medians are statistically significant. To identify though which among the income bracket groups have different median daily cigarette consumption, we need to perform the Conover-Ivan post-hoc test after rejecting the null hypothesis of a Kruskal-Wallis test.

```python
cig_cons_per_educ = [elem_smoke, hs_smoke, college_smoke, 
                     post_secondary_smoke, no_schooling]
post_hoc_educ = sp.posthoc_conover(data, val_col='daily_cigarette_intake', 
                                   group_col='highest_educ_attainment', 
                                   p_adjust='holm')
sns.heatmap(post_hoc_educ, annot=True, fmt='.2f', cmap='OrRd_r')
plt.title('Figure 12: Heatmap of the p-values for the Pairwise Comparisons '
          'Across Educational Levels\n\n', 
          fontdict= {'fontsize': 12, 'fontweight':'bold'});
```

Integrating the results of the of the EDA and the post-hoc test, we can confirm that the median amount of cigarettes consumed by those smokers that have no formal schooling at all or those that have only taken or graduated up to elementary school is lower relative to those smokers that have entered high school.

Results of the Kruskal-Wallis test confirm that the samples between some of the medians are statistically significant. Based on the plot above, we see that the median consumption of the smokers whose highest educational attainment is `no formal schooling or undisclosed` and `elementary undergrad or grad` exhibited the largest statistical difference. [Kostova et al. (2013)](#kostova) has posited that education and income are linked together and thus may explain the similarities between mean consumption of lower-income and less-educated individuals. Among the more-educated - at least from `HS undergrad or grad` to `post secondary undergrad or grad` -  we observed less to no statistical difference in terms of mean consumption. Interestingly, the same study suggested that smoking probability decreases with higher socioeconomic status in lower-to-middle income countries, but increases among those who already smoke.

**Logistic regression**

Logistic regression was done to determine what factors influence a smoker to try quitting. Before we perform logistic regression, we first have to check whether the independent variables that we are fitting into the model exhibit some degree of multicollinearity.

To do this, we calculate the VIFs and check whether a feature is correlated with other variables in the model.


```python
# get data that would be used in the logistic regression model

new_data = df[['tried_stop_smoking', '2015', 
               'seen_health_warnings_on_packaging', 'rural', 'female', 
               'smoker_count_in_hh', 'highest_educ_attainment', 
               'income_bracket', 'age', 'quit_magazine', 
               'quit_tv', 'quit_radio', 'quit_billboard', 'quit_trains',
               'quit_cinema', 'quit_hospitals', 'quit_malls', 
               'work_outside_home', 'cigarette_price']]
new_data = new_data.dropna(subset=['tried_stop_smoking'])
new_data['cigarette_price'] = (new_data['cigarette_price']
                               .fillna(new_data['cigarette_price'].mean()))

# label-encode all binary variables
rep = {'highest_educ_attainment':
       {'no formal schooling': 'no formal schooling or undisclosed',
        'no grade completed': 'no formal schooling or undisclosed',
        'refused': 'no formal schooling or undisclosed',
        'elementary undergraduate': 'elementary undergrad or grad',
        'elementary graduate': 'elementary undergrad or grad',
        'high school graduate': 'HS undergrad or grad',
        'high school undergraduate': 'HS undergrad or grad',
        'post secondary': 'post secondary undergrad or grad',
        'post secondary, includes in yrs 1, 2, or 3': \
        'post secondary undergrad or grad',
        'college undergraduate': 'college undergrad or grad',
        'college graduate': 'college undergrad or grad',
        'post graduate degree completed': \
        'post secondary undergrad or grad'},
        'income_bracket':
        {'no income': 'no income or undisclosed',
         'refused': 'no income or undisclosed',
         "don't know": 'no income or undisclosed'},
         'last_cig_bought': {'reams': 'packs', 'cigarettes': 'sticks',
                             'never bought cigarettes': 'refused to say',
                              np.nan: 'refused to say',
                             'refused': 'refused to say'},
         'quit_magazine': {'not applicable': 0, 'no': 0, 'refused': 0,
                           'yes': 1},
         'quit_tv': {'not applicable': 0, 'no': 0, 'refused': 0,
                     'yes': 1},
         'quit_radio': {'not applicable': 0, 'no': 0, 'refused': 0,
                        'yes': 1},
         'quit_billboard': {'not applicable': 0, 'no': 0, 'refused': 0,
                           'yes': 1},
         'quit_trains': {'not applicable': 0, 'no': 0, 'refused': 0,
                         'yes': 1},
         'quit_cinema': {'not applicable': 0, 'no': 0, 'refused': 0,
                         'yes': 1},
         'quit_hospitals': {'not applicable': 0, 'no': 0, 'refused': 0,
                           'yes': 1},
         'quit_malls': {'not applicable': 0, 'no': 0, 'refused': 0,
                           'yes': 1},
         'seen_health_warnings_on_packaging': \
         {'didn\'t see any cig packages': 0, 'no': 0, 'refused': 0, 'yes': 1},
         'work_outside_home': \
         {'yes': 1, 'no/don\'t work': 0, 'refused': 0}}
new_data = new_data.replace(rep)
new_data['graphic_health_warnings_2015'] = \
new_data['2015'] * new_data['seen_health_warnings_on_packaging']

# one-hot encode all categorical variables
educ_dummies = (pd.get_dummies(new_data['highest_educ_attainment'])
                .drop(['no formal schooling or undisclosed'], axis=1))
income_bracket = (pd.get_dummies(new_data['income_bracket'])
                  .drop(['no income or undisclosed'], axis=1))
new_data = new_data.drop(['highest_educ_attainment', 'income_bracket',
                          '2015', 'seen_health_warnings_on_packaging'], axis=1)


# merge one hot encoded variables with the original dataset
new_data = pd.merge(new_data, educ_dummies, how='left',
                    left_index=True, right_index=True)
new_data = pd.merge(new_data, income_bracket, how='left',
                    left_index=True, right_index=True)
```

```python
def calc_vif(X):
    """Check whether there is multicollinearity present among the features."""
    vif = pd.DataFrame()
    vif["variables"] = X.columns[1: ]
    vif["VIF"] = [variance_inflation_factor(X.values[1: ], i) 
                  for i in range(1, X.shape[1])]
    return vif

calc_vif(new_data)
```

Looking at the VIFs, there are no variables that went beyond the VIF threshold of 10 which serves as an indicator for the presence of multicollinearity. Since the data is compliant to the statistical model's assumptions, we can proceed with the implementation of the logistic regression model.

```python
model = sm.Logit(new_data['tried_stop_smoking'], 
                 sm.add_constant(new_data.drop(['tried_stop_smoking'], 
                                               axis=1))).fit()
model.summary()
```

Results of the logistic regression confirmed that a smoker’s propensity to quit is influenced by smokers noticing information about the dangers of smoking/encouraging quitting shown in `TV`, `magazine`, and health care facilities such as `hospitals`. These variables are statistically significant at an alpha level of 0.05, and have an increasing effect (due to their positive sign) on the probability of encouraging smokers to try stop smoking. Hence, it is recommended to continue using these media in launching campaigns that encourage smokers to try quitting smoking.

The statistically significant variables among highest educational attainment are `hs undergrad or grad`, `college undergrad or grad`, `elementary undergrad or grad`, `post secondary undergrad or grad`, spanning almost all the educational attainment buckets in the survey. From this, smokers that have high levels of educational attainment tend to have an increasing effect on the probability of encouraging them to try quitting smoking.

Meanwhile, the statistically significant variables in terms of income were those that belong to the lower income brackets (`1 to 3,499` and `3,500 to 4,999`) and those who earn to the `21,000 or higher` income per month. Smokers with higher income levels tend to have an increasing effect on the probability of doing smoking cessation. In addition to this, smokers that are living in `rural` areas tend to have an increasing effect on the probability that they would attempt to quit smoking.

On the other hand, smokers that tend to work outdoors along with those that live with a lot of smokers in their household tend to have a declining effect on the probability of them trying to quit smoking.

It is also interesting to see that a peso increase in cigarette prices tend to have no effect on encouraging smokers to try quitting smoking. Likewise, the interaction term between the year `2015` and `seen_health_warnings_on_packaging` as defined by `graphic_health_warnings_2015` did not generate a statistically significant p-value (although it is positive in sign). From this result, we can argue that the imposition of graphic health warnings on cigarette packaging (as mandated by the Graphic Health Warnings Law of 2013) does not generate an effect that encourages current smokers to try quitting smoking. Inferring why this is the case, we argue that most smokers tend to buy cigarettes in sticks and not in cigarette packs. When asked about the form of cigarettes that they bought, 70% of the respondents mentioned that they bought cigarettes in sticks. With this, smokers that purchase cigarettes in sticks tend to not see the graphic health warnings in cigarette packaging.

While the coefficients of the logistic regression model are useful to identify whether a feature tends to have an increasing or decreasing effect on the probability of trying to quit smoking, interpreting the coefficients in a logistic regression is not straightforward unlike in the linear regression case. With this, we would be using odds ratios as these are more interpretable in the same way as linear regression. The odds ratio is defined as the constant effect of an explanatory variable, on the likelihood that one outcome (which is usually the positive class) would occur.

**Interpretation of the odds ratio**

The odds ratio also showed interesting insights regarding a smoker’s potential to quit. Some notable highlights include the following:

- `Females` are more likely to quit smoking. Being female is associated with a 28% increase in odds of quitting smoking. Results from neuroimaging studies suggest that reward pathways are activated more in men compared to women, where men smoke due to the reinforcing effect of nicotine, while women’s cigarette consumption is related to mood regulation or in response to cigarette-related cues ([National Institute of Drug Abuse, 2020)](#gender). In fact, females with high motivation to quit experienced greater startle magnitude, or a greater defensive response to a cigarette stimulus [(Munoz, Idrissi, Sanchez-Barrera, Fernandez, and Villa, 2011)](#munoz). This means that while women find it more difficult to quit smoking than men, they do so in response to social pressures and expectations [(Smith, Bessette, Weinberger, Sheffer, and McKee, 2016)](#smith).
- The odds of quitting increases after noticing health warnings on `TV` (30%), magazines (21%), healthcare facilities such as `hospitals` (26%). It may be worthwhile to focus efforts on educational campaigns about the hazards of smoking using TV commercials, magazine advertisements, and in healthcare facilities. There is evidence that each of the health warnings in different forms of media complement each other and can positively influence smoking cessation [(Brennan, Durkin, Cotter, Harper, and Wakefield, 2011)](#brennan). It is important to note, however, that smoking *initiation* is a different behavior to advertise and measure than smoking *cessation*; the latter can be gradually provided interventions, while the former can “rapidly change over a short time span” [(Pierce,White, and Emery, 2011)](#pierce).
- Smokers who have educational attainment are at least twice as likely to quit. Among Filipino undergraduate smokers in a local university, smoking initiation and cessation was linked to high levels of anxiety which may subside beyond graduation [(Florendo, 2014)](#florendo).
- Smokers earning `21,000 and above` have twice more odds of quitting relative to smokers earning less. This is despite the price-sensitivity of lower income smokers  [(Choi, 2016)](#choi). 
- Variables which have an odds ratio less than 1 indicates negative association. Working outside the home decreases a smoker’s odds of quitting by 27%.  Having more smokers inside a household also decreases the odds of quitting by 17%. 

# Conclusion

This study evaluated the effect of graphic health warnings in reducing cigarette consumption among Filipinos by seeking to detect significant statistical differences on cigarette consumption between the years 2009 and 2015, in income classifications, and in educational attainment. We have demonstrated the use of different statistical tests to confirm or disconfirm our hypotheses using Two Proportion Z-Test, Mann-Whitney U test, Kruskal-Wallis H-test, Conover-Ivan Post-Hoc Test, and Logistic Regression, and have validated the insights by referencing current literature whenever applicable. We conclude that the propensity to quit is influenced by smokers noticing information about the dangers of smoking/encouraging quitting shown in `TV`, `magazines`, and health care facilities such as `hospitals`. Hence, encouraging smokers to quit may be continued using these media while ensuring that the messages in each medium complements each other. Educational attainment and income also appears to influence smoke cessation, so reaching out to and providing targeted interventions towards those with lower access socioeconomic statuses may help them to quit. 

It is interesting to point out that since most smokers tend to purchase cigarettes by stick and not by pack, they tend to not see the graphic health warnings that are displayed on cigarette packaging. With this, it is important for policy makers to consider putting graphic health warnings on stores along with other points of interest (e.g. public transport terminals, wet markets, etc.) so that smokers would see them more and think twice before trying to stop smoking. Likewise, it is also important to bring the campaign of smoking cessation down to the barangay or subdivision level in order to reduce the number of cigarette smokers within households. 

Finally, as cigarette consumption affects our health outcomes, deaths and diseases related to it are entirely preventable. Insights gathered from this report may aid in lawmaking to provide progressive tobacco tax policies that discourage its use for everyone’s benefit.
