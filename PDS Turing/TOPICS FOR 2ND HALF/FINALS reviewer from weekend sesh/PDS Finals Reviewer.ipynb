Let arr be a 2D ndarray constructed from `np.random.rand(10,5)`. <br>
Which of the following statements are true regarding `arr.sum(axis=1)?`

1. The result is the row-wise sum of arr.
2. The result is the column-wise sum of arr.
3. The operation flattens the rows of the array.
4. The operation flattens the columns of the arry.
5. The result has a shape of (10,).
6. The result has a shape of (5,).

````python
arr = np.random.rand(10,5)
arr
````

axis = 0:
- collapses axis 0 (flattens the rows)
- adds each items in a column to flatten the rows (column-wise sum)
- result has shape of 5 (left with column dim)

axis = 1:
- collapses axis 1 (flattens the columns)
- adds each items in a row to flatten the columns (row-wise sum)
- result has shape of 10 (left with row dim)

````python
arr = np.random.rand(10,5)
display(arr.shape)
display(arr.sum(axis=0).shape)
# display(arr.sum(axis=1).shape)
````
