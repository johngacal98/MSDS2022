## <span style="color:#33FFBB;font-size:16pt"> Masking

**Make a mask that gets all numbers divisible by 7.**

```python
mask = arr[(arr % 7 == 0)]
mask
```

**Apply the mask to the array to get all numberes divisible by 7**

## <span style="color:#33FFBB;font-size:16pt"> Masking with np.where

**Get mask using np.where**

```python
wer = np.where(arr % 7 == 0) #returns indices where it is true
wer
```

```python
arr[wer]
```

**Change all the lucky numbers divisible by 7 to 0**

```python
arr_div7 = arr.copy()
arr_div7[wer] = 0
arr_div7
```
