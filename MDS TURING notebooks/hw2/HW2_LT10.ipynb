For this homework, we will be using the following datasets: `fit_data1.csv`, `fit_data2.csv`, `fit_data3.csv`, `fit_data4.csv`.

#### Q. Load the `fit_data1.csv` dataset using pandas and print the dataframe.

```python
# Insert code here
df_data1 = pd.read_csv('fit_data1.csv')
df_data1
```

#### Q. Create two variables `x` and `y` and assign the corresponding dataframe columns to them. Convert `x` and `y` into numpy arrays.

```python
# Insert code here
x_1 = df_data1['x'] 
y_1 = df_data1['y']
x_1 = x_1.to_numpy()
y_1 = y_1.to_numpy()
```

#### Q. Plot a graph of the data as a <b><u>dashed blue line</u></b>.

```python
# Insert code here
plt.plot(x_1, y_1, 'b--')
plt.title("Dataset 1")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Q. Based on the plot above, choose an "appropriate" function to fit the data. Print the optimal parameters and SSR. 

```python
# Insert code here
def sine_linear(x_1, A, F, C, D, E):
    return A * np.sin((2 * np.pi * F * x_1) + C) + (D*x_1 + E)

popt, pcov = curve_fit(sine_linear, x_1, y_1)

ssr = sum((y_1 - sine_linear(x_1, *popt))**2)

print('A:',popt[0])
print('F:',popt[1])
print('C:',popt[2])
print('D:',popt[3])
print('E:',popt[4])

print('SSR:', ssr)
```

#### Q. Plot the data as <u><b>blue dots</u></b> and plot you fitted function as a <b><u>smooth solid red line</u></b>.

```python
# Insert code here
x_range = np.linspace(min(x_1), max(x_1), 100)

plt.plot(x_range, sine_linear(x_range, *popt), 'r-')
plt.scatter(x_1, y_1)
plt.title("Dataset 1")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Q. Based on the plot above, choose an "appropriate" function to fit the data. Print the optimal parameters and SSR.

```python
# Insert code here
def exp_power(x_2, p, q, r, s):
    return (p * np.exp(-q * x_2)) / (r * (np.power(x_2, -s)))

np.seterr(divide='ignore')
popt, pcov = curve_fit(exp_power, x_2, y_2)

ssr = sum((y_2 - exp_power(x_2, *popt))**2)

print('p:',popt[0])
print('q:',popt[1])
print('r:',popt[2])
print('s:',popt[3])
print('SSR:', ssr)
```

```python
# Insert code here
def combo(x_4,d,e,f,g,v,w,z,u,t,h,i,j): 
     return ((d * x_4**3) + (e * x_4**2) + (f * x_4) + g)*(v/(1 + np.exp(-w * (x_4 - z))))+(1/np.sqrt(2 * np.pi *u**2)) * np.exp(-(x_4 - t)**2/(2*u**2))/(h * np.sin((2 * np.pi * i * x_4) + j))
    
popt,pcov = curve_fit(combo, x_4, y_4)

ssr = sum((y_4 - combo(x_4, *popt))**2)

print('d:',popt[0])
print('e:',popt[1])
print('f:',popt[2])
print('g:',popt[3])
print('v:',popt[4])
print('w:',popt[5])
print('z:',popt[6])
print('u:',popt[7])
print('t:',popt[8])
print('h:',popt[9])
print('i:',popt[10])
print('j:',popt[11])
print('SSR:', ssr)
```
