#### Q. Create two functions called `f(x)` and `g(x)` as defined above.

```python
# Insert code here
def f(x):
    return np.sin(2*np.pi*x)

def g(x):
    return np.cos(x/np.pi)
```

```python
# Insert code here
y_1 = f(x)
y_2 = g(x)

plt.plot(x,y_1,'b-')
plt.plot(x,y_2,'r-')

plt.show()
```
